```python
import numpy as np
```

```python
n_values = [10, 100, 1000, 10000, 100000]

for dtype in [np.float32, np.float64]:
    print(f'\nPrecision: {dtype.__name__}')

    print(f"{'Value':>6} | {'Result':<16} | {'Rel. Error':<10}")

    for n_value in n_values:
        total = dtype(0.0)
        
        for num in range(n_value):
            total += dtype(0.1)

        expected = n_value * 0.1
        relative_error = abs(float(total) - expected) / expected

        print(f"{n_value:>6d} | {float(total):16.10f} | {relative_error:10.2e}")
```
