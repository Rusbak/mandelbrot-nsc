```python
import numpy as np
```

```python
def quadratic_naive(a, b, c):
    t = type(a)
    discriminant = t(np.sqrt(b * b - t(4) * a * c))
    x1 = (-b + discriminant) / (t(2)*a) 
    x2 = (-b - discriminant) / (t(2)*a) 

    return x1, x2

# example equation: x^2 - 10000.0001*x + 1 = 0
root_1 = 10000 # 1e+4
root_2 = 0.0001 # 1e-4

for dtype in [np.float16, np.float32, np.float64]:
    a, b, c = dtype(1.0), dtype(-10000.0001), dtype(1.0)
    x1, x2 = quadratic_naive(a, b, c)

    print(f'Precision: {dtype.__name__}')
    print(f"{'Root':<4} | {'Value':<7} | {'Correct':<7}")
    print(f"{'X1':<4} | {float(x1):7.1e} | {(float(x1) == root_1):7d}")
    print(f"{'X2':<4} | {float(x2):7.1e} | {(float(x2) == root_2):7d}")
```

```python
def quadratic_stable(a, b, c):
    t = type(a)
    discriminant = t(np.sqrt(b*b - t(4)*a*c))

    if b > 0:
        x1 = (-b - discriminant) / (t(2)*a)
    else:
        x1 = (-b + discriminant) / (t(2)*a)
    
    x2 = c / (a*x1) # Vieta's formula

    return x1, x2

true_small = 1.0 / 10000.0001

for dtype in [np.float16, np.float32, np.float64]:
    a, b, c = dtype(1.0), dtype(-10000.0001), dtype(1.0)

    x1_naive, x2_naive = quadratic_naive(a, b, c)
    x1_stable, x2_stable = quadratic_stable(a, b, c)

    error_naive = abs(float(x2_naive) - true_small) / true_small
    error_stable = abs(float(x2_stable) - true_small) / true_small

    print(f'Precision: {dtype.__name__}')
    print(f'Naive Error: {error_naive:.3e}')
    print(f'Stable Error: {error_stable:.3e}\n')
```
