```python
import numpy as np
```

```python
def find_machine_epsilon(dtype=np.float64):
    epsilon = dtype(1.0)

    while dtype(1.0) + epsilon / dtype(2.0) != dtype(1.0):
        epsilon /= dtype(2.0)
    return epsilon

for dtype in [np.float16, np.float32, np.float64]:
    computed = find_machine_epsilon(dtype)
    reference = np.finfo(dtype).eps

    print(f'Precision: {dtype.__name__}')
    print(f'Computed: {float(computed):.5e}')
    print(f'np.finfo: {float(reference):.5e}')
    print(f'Matches: {computed == reference}\n')
```
